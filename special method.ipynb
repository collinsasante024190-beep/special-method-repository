```python
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)
    def __sub__(self, other):
        return Vector(self.x - other.x, self.y - other.y)
    def __mul__(self, other):
        return Vector(self.x * other.x, self.y * other.y)
    def __truediv__(self, other):
        return Vector(self.x / other.x, self.y / other.y)

    def __str__(self):
        return f'answer: ({self.x}, {self.y})'

vector1 = Vector(2,4)
vector2 = Vector(3,4)
print(vector1 + vector2)
print(vector1 - vector2)
print(vector1 * vector2)
print(vector1 / vector2)


```

```python
class Programmes:
    def __init__(self):
        self.programmes = ['Calculus','French','Basic electronics','OOP']

    def __contains__(self, programmes):
        return programmes in self.programmes

p = Programmes()
print('OOP' in p)
print('Maths' in p)
```

```python
class Divider:
    def __init__(self,number):
        self.number = number

    def __call__(self,number):
        return number // self.number

div = Divider(2)
print(div(10))
```

```python
class Odd_numbers:
    def __init__(self,odd):
        self.odd = odd
        self.odd_number = 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.odd_number > self.odd:
            raise StopIteration

        value = self.odd_number
        self.odd_number += 2
        return value

    for num in Odd_numbers(10):
        print(num)

```
